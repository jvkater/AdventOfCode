```python
inp = 277678
```

```python
right_corner = [x**2 for x in range(inp) if x%2 != 0]
```

```python
right_corner
```

```python
for i in right_corner:
    if i>inp:
        rcind = i
        break
```

```python
from math import sqrt
```

```python
hor_steps = sqrt(rcind)
ths = right_corner.index(rcind)
```

```python
rdc = rcind
ldc = rcind - hor_steps+1
luc = rcind - 2*hor_steps+2
ruc = rcind - 3*hor_steps+3
```

```python
if inp>ldc:
    print ("Lower section")
if inp>luc and inp<ldc:
    print ("Left side")
if inp>ruc and inp<luc:
    print ("Top side")
if inp<ruc:
    print("Right side")
```

```python
mid_sec = rdc - (rdc-ldc)/2
mid_sec
```

```python
ver_steps = inp - mid_sec
```

```python
steps = ths+ver_steps
```
