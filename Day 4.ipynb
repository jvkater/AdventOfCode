```python
def wordcheck(tx1:str, tx2:str):
    # 
    c = 0
    #    
    for i in tx1:
        #    .   -  +1
        if i in tx2:
            c+=1
        else:
            break
    #   ,     ,  
    if c == len(tx1):
        return 'same'
    else:
        return 'dif'
```

```python
def str_chk(tx):  
    #  
    str_vdty = 'valid'
    # -   .     ,  -1
    for st in range(len(tx)-1):
        #    -   (),       
        a, *rest = tx[st:len(tx)]
        #       
        for i in rest:
            if len(a)==len(i):
                #  ,    .    -      
                #
                if wordcheck(a,i) == 'same' and wordcheck(i,a)=='same':
                    str_vdty ='not valid'
                    break
    if str_vdty == 'not valid':
        #print('string not valid')
        return 0
    else:
        #print('string is valid')
        return 1
```

```python
nr = 0
for i in arr:
    nr += str_chk(i)
```

```python
def worddict(txt1,txt2):
    d1 = dict.fromkeys(set(txt1))
    d2 = dict.fromkeys(set(txt2))
    for i in d1:
        d1[i]=txt1.count(i)
    for i in d2:
        d2[i]=txt2.count(i)
    if d1==d2:
        return 'strings are the same!'
    else:
        return 'strings are not the same!'
```

```python
worddict('abcde','edcba')
```
