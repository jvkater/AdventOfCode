```python
def func(inp_comm):
    x=inp_comm.split()
    
    if x[1]=='inc':
        d[x[0]]=d.get(x[0])+int(x[2])
    else:
        d[x[0]]=d.get(x[0])-int(x[2])
    return d[x[0]]
```

```python
def condition(inp_cond):
    x = inp_cond.split()
    if x[5] == '<':
        return d[x[4]]<int(x[6])
    if x[5] == '<=':
        return d[x[4]]<=int(x[6])
    if x[5] == '==':
        return d[x[4]]==int(x[6])
    if x[5] == '>':
        return d[x[4]]>int(x[6])
    if x[5] == '>=':
        return d[x[4]]>=int(x[6])
    if x[5] == '!=':
        return d[x[4]]!=int(x[6])
```
